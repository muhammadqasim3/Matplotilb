# Matplotlib Object Oriented Method
The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.7, 0.7]) # left, bottom, width, height (range 0 to 1)
axes.plot(x,y,'g*--')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('My Plot');
```

```python
# creating blank canvas
fig = plt.figure()

axes_main = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #outer Axes
axes_inner = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #inner Axes

#Outer Axes
axes_main.plot(x,y,'r')
axes_main.set_xlabel('x-axis')
axes_main.set_ylabel('y-axis')
axes_main.set_title('Outer Plot')
#Inner Axes
axes_inner.plot(x,y,'g')
axes_inner.set_xlabel('x-axis')
axes_inner.set_ylabel('y-axis')
axes_inner.set_title('Inner Plot');
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
fig, axes = plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

'''We can use fig.tight_layout() or plt.tight_layout() method, which automatically adjusts the positions of the axes on the figure 
canvas so that there is no overlapping content:'''    
plt.tight_layout()
# Display the figure object    
fig
```

### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
# same arguments can also be passed to layout managers, such as the subplots function:
fig, axes = plt.subplots(figsize=(9,3)) # width, height

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Plot');
```

### Legends

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")
axes.legend(loc=0);
```

## Setting colors, linewidths, linetypes

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b.-')
axes.plot(x,x, 'g.--');
```

```python
fig, axes = plt.subplots()

axes.plot(x, x, color="blue", alpha=0.5) # half-transparant
axes.plot(x,x+1, color="pink")
axes.plot(x, x+2, color="orange")
axes.plot(x, x+3, color="green");       
```

### Line and marker styles

```python
fig, axes = plt.subplots(figsize=(18,10))

axes.plot(x, x, color="red", linewidth=0.25)
axes.plot(x,x+1,color="purple", linewidth = 0.50)
axes.plot(x,x+2,color="orange", linewidth = 1.0)
axes.plot(x,x+3,color="green", linewidth = 2.0)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
axes.plot(x, x+4, color="purple", lw=3, linestyle='-')
axes.plot(x, x+5, color="green", lw=3, ls='-.')
axes.plot(x, x+6, color="blue", lw=3, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
axes.plot(x, x+7, color="brown", lw=3, ls='-', marker='+')
axes.plot(x, x+8, color="green", lw=3, ls='--', marker='o')
axes.plot(x, x+9, color="orange", lw=3, ls='-', marker='s')
axes.plot(x, x+10, color="purple", lw=3, ls='--', marker='1')

# marker size and color
axes.plot(x, x+11, color="green", lw=1, ls='-', marker='o', markersize=2)
axes.plot(x, x+12, color="blue", lw=1, ls='-', marker='o', markersize=4)
axes.plot(x, x+13, color="brown", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x, x+14, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Plot range

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
